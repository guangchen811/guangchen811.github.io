```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the SDE
a = 2.0
b = 1.0

# Define the time step and the number of time steps
dt = 0.01
T = 1.0
N = int(T / dt)

# Initialize the state variable and the Wiener process
X = np.zeros(N+1)
W = np.zeros(N+1)

# Simulate the Wiener process
for i in range(N):
    W[i+1] = W[i] + np.sqrt(dt) * np.random.normal()

# Simulate the SDE using the Euler-Maruyama method
for i in range(N):
    X[i+1] = X[i] + a*X[i]*dt + b*X[i]*np.sqrt(dt)*W[i+1]

# Plot the results
plt.plot(np.arange(0, T+dt, dt), X)
plt.xlabel('Time')
plt.ylabel('X')
plt.title('Simulated SDE')
plt.show()
```
